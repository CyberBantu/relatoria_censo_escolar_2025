```python
from pathlib import Path
import pandas as pd

base_dir = Path("dados")
csv_files = list(base_dir.glob("*.csv"))

print("Pasta usada:", base_dir.resolve())
print("Arquivos encontrados:", [f.name for f in csv_files])

for csv_path in csv_files:
    df = pd.read_csv(csv_path, sep=";", encoding="latin-1", low_memory=False)
    parquet_path = csv_path.with_suffix(".parquet")
    df.to_parquet(parquet_path, index=False)

print(f"{len(csv_files)} arquivo(s) CSV convertido(s) para Parquet em: {base_dir.resolve()}")
```
